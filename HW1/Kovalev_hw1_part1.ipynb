## Task 1. Numpy Problem 1 (1 point)

Write a function, which takes a matrix `arr` and centers each of its rows
by the mean within that row. Check [this](https://docs.scipy.org/doc/numpy/reference/routines.statistics.html)
out for documentation.

```python
def center(arr):
    """Center each row of a matrix by the mean value in that row.

    Parameters
    ----------
    arr : array, shape = (n_rows, n_cols)
        The matrix, the rows of which to center.

    Returns
    ----------
    out : array, shape = (n_rows, n_cols)
        The final row-centered matrix.
    """
    assert arr.ndim == 2
    n_rows, n_cols = arr.shape

    ### BEGIN Solution
    out = arr - arr.mean(1).reshape(-1, 1)
    ### END Solution

    return out
```

## Task 2. Numpy Problem 2 (1 point)

Implement the following fancy function:
\begin{equation}
    f(x) = \sigma\bigl(
                \max\{x + 5, 0\} + \max\{5 - x, 0\}
                + \max\{\min\{\cos(2 x \pi), \tfrac12\}, -\tfrac14\}
            \bigr)
         \,,
\end{equation}
where $\sigma(x) = (1+e^{-x})^{-1}$ is the sigmoid function.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def fancy_function(x):
    """Compute some fancy function.
    
    Parameters
    ----------
    x : array, 1 dimendional, shape=(n_samples,)
        The array argument values.

    Returns
    -------
    y : array, 1 dimendional, shape=(n_samples,)
        The values of the fancy function.
    """
    ### BEGIN Solution
    out = sigmoid(np.maximum(x + 5, 0) + np.maximum(5 - x, 0) + np.maximum(np.minimum(np.cos(2 * x * np.pi), 0.5), -0.25))
    ### END Solution
    
    return out
```

### Subproblem 4.2. Pipeline: Fitting to Data by Feature Engeneering (1 of 2 points)
In previous task 1.1 several classifiers obviously failed fitting to data. This happend because the decision rule of the classifier has a restricted form (e.g. linear for linear models), while the data is more complicated.

One may try to change the parameters of the classifier (e.g. increase the number of trees in Forest) in order to improve accuracy, but some models (especially linear) do not have parameters that can change the form of the decision rule.

In this case, the **feature engeneering** helps: one may try to compute new (e.g. non-linear) features based on the existing pool and fit the classifier in the new features. This may help low-complex claffifiers to fit to hard data dependencies.

Your task it to
* Choose **two classifiers** from the following: Decision Tree, Random Forest, Naive Bayes, Logistic Regression;
* By generating of additional features (e.g. polynomial) make them achieve accuracy $>0.95$.
* For each classifier, write 2-3 sentences why did you decide to choose these features.

It is your choice how to generate features. You may create hand-crafted featues and add them manualy. Nevertheless, we **highly suggest** to get used to and apply the following builtin `sklearn` methods:
* `PolynomialFeatures`, `GaussianRandomProjection` among others - for feature generation
* `StandartScaler`, `MinMaxScaler` among others - for feature scaling
* `Pipeline` - for combining several operations in a row (e.g. feature creation & prediction)

```python
### BEGIN Solution
fig, axes = plt.subplots(
    ncols=2,
    nrows=1, figsize=(21, 12),
    dpi=75
)


poly_dt = PolynomialFeatures(degree=2, include_bias=False)
dt = DecisionTreeClassifier(max_depth=4)
pp_dt = Pipeline([('polynomial features', poly_dt), ('decision tree', dt)])
pp_dt.fit(X, y)
plot_decision_regions(X=X, y=y, clf=pp_dt, ax=axes[0], legend=2)
axes[0].set_title('DecisionTreeClassifier, accuracy = {:0.3f}'.format(accuracy_score(y, pp_dt.predict(X))))

poly_lr = PolynomialFeatures(degree=5, include_bias=False)
lr = LogisticRegression()
pp_lr = Pipeline([('polynomial features', poly_lr), ('logistic regression', lr)])
pp_lr.fit(X, y)
plot_decision_regions(X=X[:, :2], y=y, clf=pp_lr, ax=axes[1], legend=2)
axes[1].set_title('LogisticRegression, accuracy = {:0.3f}'.format(accuracy_score(y, pp_lr.predict(X))))
plt.show()
### END Solution
```

As can be seen from the pictures in 4.1, simple decision tree with small depth lacks an ability to capture complex non-linear patterns in data. However, it is shown that the border between two classes is definitely not a line, it looks more like a polynomial of some power. Therefore my decision was to generate polynomial features, which gave an increase in resulting accuracy. Analogically, we can see that basic logistic regression is drawing a line as a border. To help it capture more non-linearity, I added polynomial features, which also worked an gave a boost in quality.

I decided to tune number of trees in the forests, maximum depth of a tree, maximum number of features to consider when splitting and minimum number of samples in a leaf.

```python
### BEGIN Solution
rf = RandomForestRegressor(random_state=13)
params = {
    'n_estimators': [1, 5, 10, 20, 40, 75, 100, 150, 200],
    'max_depth': np.arange(3, 11),
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_leaf': [1, 5, 10, 15]
}
gridcv_rf = GridSearchCV(rf, params, cv=5, scoring='neg_mean_squared_error')
gridcv_rf.fit(X_train, y_train)

rf = gridcv_rf.best_estimator_
rf.fit(X_train, y_train)

rf_train = mean_squared_error(y_train, rf.predict(X_train))
rf_val = -gridcv_rf.best_score_
rf_test = mean_squared_error(y_test, rf.predict(X_test))

et = ExtraTreesRegressor(random_state=13)
params = {
    'n_estimators': [1, 5, 10, 20, 40, 75, 100, 150, 200],
    'max_depth': np.arange(3, 11),
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_leaf': [1, 5, 10, 15]
}
gridcv_et = GridSearchCV(et, params, cv=5, scoring='neg_mean_squared_error')
gridcv_et.fit(X_train, y_train)

et = gridcv_et.best_estimator_
et.fit(X_train, y_train)

et_train = mean_squared_error(y_train, et.predict(X_train))
et_val = -gridcv_et.best_score_
et_test = mean_squared_error(y_test, et.predict(X_test))

res = pd.DataFrame(data=[
    ['ExtraTreesRegressor', et_train.round(2), et_val.round(2), et_test.round(2)],
    ['RandomForestRegressor', rf_train.round(2), rf_val.round(2), rf_test.round(2)]
], columns=['Ensemble Classifier', 'Train', 'Valid', 'Test'])
print(res.set_index('Ensemble Classifier'))
### END Solution
```

Forests can easily overfit on data due to their complex structure. So here it is clearly shown - the quality of both algorithms on train is significantly better than on validation/test. This is happens because trees in this forests are relatively deep, while the amount of data is not so big.

```python
print('max_depth in Random Forest: {}'.format(gridcv_rf.best_params_['max_depth']))
print('max_depth in Extremely Randomized Forest: {}'.format(gridcv_et.best_params_['max_depth']))
```

Extremely Randomized Forest algorithm in this case performs slightly better Random Forest. Probably it happens because built decision trees in Extremely Randomized Forest tend to make less correlated mistakes than in case of Random Forest due to the extra randomness in choosing a threshold, and in average they perform better in this very problem.

## Task 7. Multi-Label Classification Strategies (2 points)
In this task you deal with multiclass classification problem for [Glass Classification Data](https://www.kaggle.com/uciml/glass). Lets load the dataset.

More precisely, for every pair (algorithm, approach) you are to perform 5-fold cross validation on the data and output the validation score and the computation time in the **table** form. As we can see, there is some imbalance between the classes. Here, let's focus on maximizing the number of correct predictions by the classifiers, and not on avoiding the errors on smaller classes. Hence, F1-score with micro-averaging is used.

```python
### BEGIN Solution

#https://stackoverflow.com/questions/38783027/jupyter-notebook-display-two-pandas-tables-side-by-side
def display_side_by_side(df1, df2):
    html_str='<p>Cross Validation Scores</p>' + df1.to_html() + '<br><p>Training Duration (CV=5)</p>' + df2.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)


multiclass_scores = []
ovr_scores = []
ovo_scores = []

multiclass_time = []
ovr_time = []
ovo_time = []

clfs = [
    LogisticRegression(),
    RandomForestClassifier(n_estimators=100, max_depth=3),
    KNeighborsClassifier(n_neighbors=5),
    DecisionTreeClassifier(max_depth=7)
]

for clf in clfs:
    clf_ovr = OneVsRestClassifier(clf)
    clf_ovo = OneVsOneClassifier(clf)
    
    start_time = time.time()
    multiclass_scores.append(cross_val_score(clf, X, y, cv=5, scoring='f1_micro').mean())
    multiclass_time.append(time.time() - start_time)
    
    start_time = time.time()
    ovr_scores.append(cross_val_score(clf_ovr, X, y, cv=5, scoring='f1_micro').mean())
    ovr_time.append(time.time() - start_time)
    
    start_time = time.time()
    ovo_scores.append(cross_val_score(clf_ovo, X, y, cv=5, scoring='f1_micro').mean())
    ovo_time.append(time.time() - start_time)

index = ['Multiclass', 'OneVsRest', 'OneVsOne']
columns = ['Logistic', 'Random Forest', 'K Nearest Neighbors', 'Decision Tree']

results = pd.DataFrame(
    data=[multiclass_scores, ovr_scores, ovo_scores],
    index=index,
    columns=columns
)

results_time = pd.DataFrame(
    data=[multiclass_time, ovr_time, ovo_time],
    index=index,
    columns=columns
)

display_side_by_side(results, results_time)
### END Solution
```

As the results show, OneVsRest approach gives a nice boost in quality comparing to multiclass approach without significant increase in computation time only in case of kNN. In all other cases, OneVsRest/OneVsOne either don't win in quality to Multiclass or need a significantly bigger amount of time to finish their job.
